```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = pd.read_csv('titanic.csv')
titanic_data.head()
```

```python
titanic_data.info()
```

```python
titanic_data['Age'].unique()
```

```python
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].mean())
```

```python
titanic_data['Age'].isnull().sum()
```

```python
titanic_data['Fare'].unique()
```

```python
titanic_data['Fare'] = titanic_data['Fare'].fillna(0)
titanic_data['Fare'].isnull().sum()
```

```python
titanic_data['Cabin'].unique()
```

```python
titanic_data['Cabin'].nunique()
```

```python
titanic_data['Cabin'] = titanic_data['Cabin'].apply(lambda x:0 if pd.isnull(x) else 1)
titanic_data['Cabin'].isna().sum()
```

```python
titanic_data.info()
```

```python
titanic_data = titanic_data.drop(['PassengerId','Name','Ticket'], axis=1)
titanic_data.head()
```

```python
titanic_data['Embarked'].unique()
```

```python
titanic_data['Embarked'] = titanic_data['Embarked'].map({'Q':1, 'S':2, 'C':3})
titanic_data['Embarked'].unique()
```

```python
titanic_data['Sex'] = titanic_data['Sex'].map({'male':0, 'female':1})
```

```python
titanic_data.info()
```

```python
titanic_data.head()
```

```python
titanic_data.describe()
```

```python
sns.catplot(x ="Sex", hue ="Survived", kind ="count", data = titanic_data)
```

```python
# Group the dataset by Pclass and Survived and then unstack them
group = titanic_data.groupby(['Pclass', 'Survived'])
pclass_survived = group.size().unstack()

# Heatmap - Color encoded 2D representation of data.
sns.heatmap(pclass_survived, annot = True, fmt ="d")
```

```python
# Histogram
plt.hist(titanic_data['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
# Divide Fare into 4 bins
titanic_data['Fare_Range'] = pd.qcut(titanic_data['Fare'], 4)

# Barplot - Shows approximate values based
# on the height of bars.
sns.barplot(x ='Fare_Range', y ='Survived', data = titanic_data)
```

```python
plt.scatter(titanic_data['Age'], titanic_data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare')
plt.show()
```

```python
sns.boxplot(x=titanic_data['Survived'], y=titanic_data['Fare'])
plt.xlabel('Survival Status')
plt.ylabel('Fare')
plt.title('Survival Status vs. Fare')
plt.show()
```
